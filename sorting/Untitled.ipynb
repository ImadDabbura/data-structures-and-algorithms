# Insertion Sort

```python
def insertion_sort(data):
    '''Sorts the array in ascending order using Insertion-Sort method.'''
    n = len(data)
    
    # Empty or single element array is already sorted.
    if n <= 1:
        return data
    
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
```

```python
l = []
insertion_sort(l)
l
```

```python
l = [2]
insertion_sort(l)
l
```

```python
l = list(range(10, 0, -1))
l
```

```python
insertion_sort(l)
l
```
