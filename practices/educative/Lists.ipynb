# Challenge 1: Remove Even Integers from List

```python
def remove_even(List):
    return [num for num in List if num % 2 != 0]
```

```python
l = [1, 2, 4, 5, 10, 6, 3]
remove_even(l)
```

# Challenge 2: Merge Two Sorted Lists

```python
def merge_arrays(lst1,lst2):
    ans = []
    i = j = 0
    while i < len(lst1) and j < len(lst2):
        if lst1[i] <= lst2[j]:
            ans.append(lst1[i])
            i += 1
        else:
            ans.append(lst2[j])
            j += 1
    if i < len(lst1):
        ans.extend(lst1[i:])
    elif j < len(lst2):
        ans.extend(lst2[j:])
    return ans
```

```python
arr1 = [1, 3, 4, 5]  
arr2 = [2, 6, 7, 8]
merge_arrays(arr1, arr2)
```

```python
merge_arrays([4,5,6],[-2,-1,0,7])
```

```python
def merge_arrays_2(lst1,lst2):
    ans = [0] * (len(lst1) + len(lst2))
    i = j = k = 0
    while i < len(lst1) and j < len(lst2):
        if lst1[i] <= lst2[j]:
            ans[k] = lst1[i]
            i += 1
        else:
            ans[k] = lst2[j]
            j += 1
        k += 1
    while i < len(lst1):
        ans[k] = lst1[i]
        i += 1
        k += 1
    while j < len(lst2):
        ans[k] = lst2[j]
        j += 1
        k += 1
    return ans
```

```python
merge_arrays_2(arr1, arr2)
```

```python
merge_arrays_2([4,5,6],[-2,-1,0,7])
```

# Challenge 3: Find Two Numbers that Add up to "n"

```python
def find_sum(lst, value):
    for i in range(len(lst) - 1):
        diff = value - lst[i]
        if diff in lst[i:]:
            return [lst[i], diff]
    return False
```

```python
l = [1,21,3,14,5,60,7,6]
n = 81
find_sum(l, n)
```

This take $O(n^2)$

```python
def find_sum(lst, value):
    found_values = set()
    for e in lst:
        diff = value - e
        if diff in found_values:
            return [e, diff]
        found_values.add(e)
    return False
```

Takes $O(n)$.

We can also sort the list and then look for values which takes $O(nlogn)$.

# Challenge 4: List of Products of all Elements

```python
def find_product(arr):
    l = [0] * len(arr)
    for i in range(len(arr)):
        ans = 1
        for j in range(len(arr)):
            if i != j:
                ans *= arr[j]
        l[i] = ans
    return l
```

```python
arr = [1, 2, 3, 4]
find_product(arr)
```

This takes $O(n^2)$.

```python
def find_product_2(arr):
    left = 1
    product = []
    
    # each element would be the product of elements before it
    for e in arr:
        product.append(left)
        left *= e
    # each product would then be multiplied by the elements after it
    right = 1
    for i in range(len(arr) - 1, -1, -1):
        product[i] *= right
        right *= arr[i]
    
    return product
```

```python
arr = [1, 2, 3, 4]
find_product_2(arr)
```

This takes $O(n)$

# Challenge 5: Find Minimum Value in List

```python
def find_minimum(arr):
    ans = arr[0]
    for e in arr[1:]:
        if e < ans:
            ans = e
    return ans
```

```python
arr = [9, 2, 3, 6]
find_minimum(arr)
```

Takes $O(n)$

# Challenge 6: First Non-Repeating Integer in a list

```python
def find_first_unique(lst):
    for i, e in enumerate(lst):
        if e not in lst[i + 1:] and e not in lst[:i]:
            return e
        
```

```python
arr = [9, 2, 3, 2, 6, 6]
find_first_unique(arr)
```

```python
arr = [2, 3, 2, 3, 6]
find_first_unique(arr)
```

Takes $O(n^2)$

```python
def find_first_unique_2(lst):
    counts = {}
    counts = counts.fromkeys(lst, 0)
    for e in lst:
        counts[e] += 1
    for e in lst:
        if counts[e] == 1:
            return e
```

```python
arr = [9, 2, 3, 2, 6, 6]
find_first_unique_2(arr)
```

```python
arr = [2, 3, 2, 3, 6]
find_first_unique_2(arr)
```

Takes $O(n)$

# Challenge 7: Find Second Maximum Value in a list

```python
def find_second_maximum(lst):
    max1 = max2 = lst[0]
    for e in lst[1:]:
        if e > max1:
            max1 = e
            max2 = max1
        elif e > max2 or max1 == max2:
            max2 = e
    return max2
```

```python
arr = [9, 2, 3, 6]
find_second_maximum(arr)
```

Takes $O(n)$

# Challenge 8: Right Rotate List

```python
def right_rotate(lst, n):
    return lst[-n:] + lst[:-n]
```

```python
lst = [1, 2, 3, 4, 5]
n = 3
right_rotate(lst, n)
```

Takes $O(n)$

# Challenge 9: Rearrange Positive & Negative Values

```python
def re_arrange(lst):
    return [e for e in lst if e < 0] + [e for e in lst if e >= 0]
```

```python
arr = [10, -1, 20, 4, 5, -9, -6]
re_arrange(arr)
```

```python
def re_arrange_2(lst):
    pos_idx = 0
    for i, e in enumerate(lst):
        if e < 0:
            if i != pos_idx:
                lst[pos_idx], lst[i] = lst[i], lst[pos_idx]
            pos_idx += 1
    return lst
```

```python
arr = [10, -1, 20, 4, 5, -9, -6]
re_arrange_2(arr)
```

Both take $O(n)$

# Challenge 10: Rearrange Sorted List in Max/Min Form

```python
def max_min(lst):
    l = []
    for i in range(len(lst) // 2):
        l.append(lst[-(i + 1)])
        l.append(lst[i])
    if len(lst) % 2:
        l.append(lst[len(lst) // 2])
    return l
```

```python
lst = [1,2,3,4,5]
max_min(lst)
```

Takes $O(n)$.
