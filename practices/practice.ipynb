```python
def remove_even(L):
    return [e for e in L if e % 2 != 0]
```

```python
l = [1, 2, 4, 5, 10, 6, 3]
remove_even(l)
```

# Q2

```python
def merge_arrays(l1, l2):
    n, m = len(l1), len(l2)
    res = [None] * (n + m)
    i = j = k = 0
    
    while i < n and j < m:
        if l1[i] <= l2[j]:
            res[k] = l1[i]
            i += 1
        else:
            res[k] = l2[j]
            j += 1
        k += 1

    if i < n:
        res[j:] = l1[i:]
    elif j < m:
        res[i:] = l2[j:]
    
    return res
```

```python
arr1 = [1, 3, 4, 5]  
arr2 = [2, 6, 7, 8]
merge_arrays(arr1, arr2)
```

# Q3

```python
def find_sum(l, value):
    n = len(l)
    for i in range(n):
        diff = value - l[i]
        if diff in l[i + 1:]:
            return[l[i], diff]
        
    return False
```

```python
l = [1,21,3,14,5,60,7,6]
n = 81
find_sum(l, n)
```

```python
def find_sum(l, value):
    found_values = set()
    for e in l:
        diff = value - e
        if diff in found_values:
            return [e, diff]
        found_values.add(e)
    return False
```

```python
l = [1,21,3,14,5,60,7,6]
n = 81
find_sum(l, n)
```

# Q4

```python
def find_product(arr):
    n = len(arr)
    res = [None] * n
    for i in range(n):
        tmp = 1
        for j in range(n):
            if i != j:
                tmp *= arr[j]
        res[i] = tmp
    return res
```

```python
arr = [1, 2, 3, 4]
find_product(arr)
```

```python
def find_product(arr):
    n = len(arr)
    ans = [None] * n
    left = 1
    for i, e in enumerate(arr):
        ans[i] = left
        left *= e
    right = 1
    for i in range(n - 1, -1, -1):
        ans[i] *= right 
        right *= arr[i]
    return ans
```

```python
arr = [1, 2, 3, 4]
find_product(arr)
```

# Q5

```python
def find_min(arr):
    ans = arr[0]
    for e in arr[1:]:
        if e < ans:
            ans = e
    return ans
```

```python
arr = [9, 2, 3, 6]
find_min(arr)
```

# Q6

```python
def find_first_uniq(l):
    for i, e in enumerate(l):
        if e not in l[:i] and e not in l[i + 1:]:
            return e
```

```python
arr = [9, 2, 3, 2, 6, 6]
find_first_uniq(arr)
```

```python
def find_first_uniq(l):
    counts = dict.fromkeys(l, 0)
    for e in l:
        counts[e] +=1
    for e in l:
        if counts[e] == 1:
            return e
```

```python
arr = [9, 2, 3, 2, 6, 6]
find_first_uniq(arr)
```

# Q7

```python
def find_second_max(l):
    max1 = max2 = l[0]
    for e in l[1:]:
        if e > max1:
            max2 = max1
            max1 = e
        elif e > max2 or max1 == max2:
            max2 = e
    return max2
```

```python
arr = [9, 2, 3, 6]
find_second_max(arr)
```

# Q9

```python
def re_arrange(l):
    return [e for e in l if e < 0] + [e for e in l if e >= 0]
```

```python
arr = [10, -1, 20, 4, 5, -9, -6]
re_arrange(arr)
```

# Q10

```python
def max_min(l):
    res = []
    n = len(l)
    for i in range(0, n // 2, 1):
        res.append(l[-(i + 1)])
        res.append(l[i])
    if n % 2 != 0:
        res.append(l[n // 2])
    return res
```

```python
lst = [1,2,3,4,5]
max_min(lst)
```
