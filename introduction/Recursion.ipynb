```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(0)
```

```python
factorial(3)
```

```python
def binary_search(data, target, low, high):
    if low > high:
        return False
    else:
        mid = (low + high) // 2
        if target == data[mid]:
            return True
        elif target > data[mid]:
            low = mid + 1
            return binary_search(data, target, low, high)
        else:
            high = mid - 1
            return binary_search(data, target, low, high)
```

```python
binary_search([1], 1, 0, 0)
```

```python
binary_search([1], 0, 0, 0)
```

```python
def binary_search_iterative(data, target):
    low = 0
    high = len(data) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if target == data[mid]:
            return True
        elif target > data[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return False
```

```python
binary_search_iterative([1], 1)
```

```python
binary_search_iterative([1], 0)
```

```python
def reverse(data, start, stop):
    if start < stop - 1:
        data[start], data[stop - 1] = data[stop - 1], data[start]
        reverse(data, start + 1, stop - 1)
```

```python
def sum_recursive(data):
    if len(data) == 1:
        return data[0]
    else:
        return data[0] + sum_recursive(data[1:])
```
