## Matching Parantheses

```python
def is_matched(exp):
    left = "({["
    right = ")}]"
    S = ResizedArrayStack()
    for char in exp:
        if char in left:
            S.push(char)
        elif char in right:
            if S.is_empty():
                return False
            elif left.index(S.pop()) != right.index(char):
                return False
    return S.is_empty()
```

```python
is_matched("()"), is_matched("[]"), is_matched("{}")
```

```python
is_matched("(")
```

```python
is_matched("{{([])}}")
```
